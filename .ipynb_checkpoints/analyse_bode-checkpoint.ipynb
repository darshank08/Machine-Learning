# Neural network Model

```python
from keras.models import Sequential
from keras.layers import BatchNormalization
from keras.layers import Dense, Activation
```

```python
def create_model(input_neurons):
  model = Sequential()
  #model.add(Dropout(0.2))
  model.add(Dense(64, input_dim=input_neurons, activation='relu'))
  model.add(Dense(128, activation='relu'))
  #keras.layers.Dropout(0.2)
  #model.add(Dropout(0.2))
  model.add(Dense(256, activation='relu'))
  model.add(Dense(512, activation='relu'))
  model.add(Dense(256, activation='relu'))
  model.add(Dense(128, activation='relu'))
  model.add(Dense(64, activation='relu'))
  #model.add(Dense(256, activation='relu'))
  model.add(Dense(1, activation='linear'))

  return model
```

```python
df_x = df[['Frequency', 'Z negative' ,'ET','Negative_Z_doubleDash','Z','negative_phase', 'time']]
df_y = df[['Concentration']]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_numeric = df_x.select_dtypes(include='number')
X_scaled = scaler.fit_transform(df_numeric)
```

```python
y_scaled = scaler.fit_transform(df_y)
```

# Splitting the dataset into training and testing set

```python
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X_scaled, y_scaled, test_size=0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

# ANN

```python
activation=tf.nn.swish
```

```python
model = create_model(6)
```

```python
model.compile(optimizer='Adam', loss='mean_squared_error')
```

```python
epochs_hist_Ethanol = model.fit(
    X_train, Y_train, epochs=550, batch_size=50, validation_split=0.2)
```

```python
epochs_hist_Ethanol.history.keys()
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt

RMSE = float(format(np.sqrt(mean_squared_error(Y_test, y_predict))))
MSE = mean_squared_error(Y_test, y_predict)
MAE = mean_absolute_error(Y_test, y_predict)
NRMSE = RMSE/np.mean(y_predict)

print('RMSE =',RMSE, '\nMSE =',MSE, '\nMAE =',MAE,'\nNRMSE =',NRMSE) 
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(Y_test, y_predict)
```
